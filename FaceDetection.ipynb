```python
import face_recognition as fr
import cv2
import numpy as np
import os

#create 2 lists that store the name of images and their face encoding
path = "./train/"
known_names = []
known_name_encodings = []
images = os.listdir(path)

#loop each image in train dr. extract name, calculate face encoding then store inf. in lists
for _ in images:
    image = fr.load_image_file(path + _)
    image_path = path + _
    encoding = fr.face_encodings(image)[0]
    known_name_encodings.append(encoding)
    known_names.append(os.path.splitext(os.path.basename(image_path))[0].capitalize())
print(known_names)

#Read the test image using the cv2 imread() method.
test_image = "./test/test.jpg"
image = cv2.imread(test_image)
# image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#face location method locate cordinates of each face detected in the test image, so find enc.
#loop each face found in the test image then compare info. with the other in train folder
face_locations = fr.face_locations(image)
face_encodings = fr.face_encodings(image, face_locations)


#draw a rectangle with the face location coordinates using methods of cv2 modules
for (top, right, bottom, left), face_encoding in zip(face_locations, face_encodings):
    matches = fr.compare_faces(known_name_encodings, face_encoding)
    name = ""

    face_distances = fr.face_distance(known_name_encodings, face_encoding)
    best_match = np.argmin(face_distances)

    if matches[best_match]:
        name = known_names[best_match]

    cv2.rectangle(image, (left, top), (right, bottom), (0, 0, 255), 2)
    cv2.rectangle(image, (left, bottom - 15), (right, bottom), (0, 0, 255), cv2.FILLED)
    font = cv2.FONT_HERSHEY_DUPLEX
    cv2.putText(image, name, (left + 6, bottom - 6), font, 1.0, (255, 255, 255), 1)

#show the final result 
cv2.imshow("Result", image)
#Save the image to our current working directory using the imwrite() method.
cv2.imwrite("./output.jpg", image)
cv2.waitKey(0)
cv2.destroyAllWindows()

```
